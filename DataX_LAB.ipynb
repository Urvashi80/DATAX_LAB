Quick Data Cleaning

1.Fill missing 'income' values

 we will use the median income, as it's less sensative to outliers.
2.clean 'marital_status' column

 we will replace 'absurd', 'YOLO', and 'Alone' with 'single' to standardize.

```python
median_income = df['Income'].median()
df['Income'] = df['Income'].fillna(median_income)
print(f"Filled {df['Income'].isnull().sum()} missing income values with the median: {median_income}")
```

```python
df['Marital_Status'] = df['Marital_Status'].replace(['Absurd', 'YOLO', 'Alone'], 'Single')
print("\nCleaned 'Marital_Status' values:")
print(df['Marital_Status'].unique())
```

```python
df = df.drop(columns=['Z_CostContact', 'Z_Revenue'])
print("\nDropped constant Z_ columns.")
```

Example 2: Filter rows on one condition

 Get all customers with an income over $100,000

```python
high_earners = df[df['Income'] > 100000]
print(f"\nFound {len(high_earners)} customers with income > $100k")
display(high_earners)
```

Example 3: Filter rows on multiple conditions

 Get customers with a 'PhD' who also have children ('Kidhome' or 'Teenhome')

```python
df['Teenhome'] = pd.to_numeric(df['Teenhome'], errors='coerce')

phd_with_kids = df[
    (df['Education'] == 'PhD') &
    ((df['Kidhome'] > 0) | (df['Teenhome'] > 0))
]
print(f"\nFound {len(phd_with_kids)} PhD customers with kids.")
```
